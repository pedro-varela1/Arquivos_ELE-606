### Perguntas – Introdução a IA usando Python

1. ***O que é IA? Defina inteligência artificial em suas próprias palavras.***

A Inteligência Artificial (IA) é um campo da ciência da computação que se dedica a criar sistemas e algoritmos que conseguem realizar tarefas que normalmente requerem inteligência humana, como aprendizado, raciocínio, percepção, linguagem natural, reconhecimento de voz e imagem, tomada de decisão e resolução de problemas complexos. Embora tenha origens na década de 1950, a IA teve um progresso notável nas últimas décadas devido a avanços em hardware, software e técnicas de aprendizado, além do aumento de dados disponíveis para treinamento. A IA tem sido aplicada em várias áreas, incluindo assistentes virtuais, veículos autônomos, análise de dados, medicina, finanças e entretenimento, causando transformações significativas na forma como vivemos e trabalhamos.

2. ***Quais são os componentes básicos de um sistema de IA?***

Como componentes básicos, podemos citar:
- ***Dados de Treinamento*** - já requer um conjunto de dados para aprender e generalizar a partir dele
- ***Algoritmos de Aprendizagem*** - permitem que o sistema de IA aprenda com os dados e melhore seu desempenho ao longo do tempo. Eles podem ser supervisionados, não supervisionados ou de aprendizado por reforço, dependendo do tipo de tarefa.
- ***Modelos de Aprendizagem*** - pode ser uma rede neural, uma árvore de decisão, um modelo estatístico, lógica fuzzy, entre outros.

Pode conter outros componentes periféricos, como uma interface com usuário e estratégias de processamento de dados de treinamento, conforme a estrutura dos dados de entrada.

3. ***Quais são algumas aplicações comuns de IA no mundo real hoje?***

Há várias áreas de aplicação da IA:
- ***Processamento de Linguagem Natural (NLP)*** - os modelos de IA podem ser empregados no processamento de texto para: análise de sentimento, geração de textos e tradução de idiomas, por exemplo.
- ***Automação Residencial*** - podemos citar os assistentes virtuais e chatbots como: Siri, Alexa e Google Assistant, que respondem perguntas, realizam agendamentos e controlam dispositivos conectados à rede da casa.
- ***Processamento de Imagens e Vídeos*** - com uso na área médica para, por exemplo, diagnosticar doenças, uso em softwares de edição para automatização desse processo e uso por veículos autônomos para reconhecimento do ambiente.
- ***Robótica*** - automatização de tarefa em fábricas.
- ***Games*** - podemos usar uma IA para gerar ambientes e enredos dinamicamente e criar personagens não jogáveis inteligentes.



4. ***O que é aprendizado de máquina? Como ele se relaciona com IA?***

Machine Learning é uma subárea da IA que está relacionado à criação de modelos que aprendam a partir de dados (IA é um conceito mais amplo, envolvendo a realização de tarefas diversas). Ao invés um ser humano estudar os dados e aplicar lógica de programação, a máquina analisa esses dados e identifica padrões.

5. ***Como os algoritmos de aprendizado de máquina "aprendem" a partir de dados?***

Há várias maneiras dos algoritmos de Machine Learning aprenderem a partir de dados. Geralmente, isso se dá com processos de otimização, ajuste de parâmetros e processos interativos. Por exemplo, o algoritmo Backpropagation realiza uma passagem para trás (backpropagation) para ajustar os parâmetros de um modelo de rede neural, visando minimizar o erro quadrático médio - a entrada passa por camadas com pesos e funções de ativação, e, ná saída, com base na comparação com a saída desejada, os pesos são ajustados e o processo ocorre novamente.


6. *** Quais são os diferentes tipos de aprendizado de máquina (supervisionado, não supervisionado, por reforço)?***

Há três tipos de Machine Learning
- ***Aprendizado Supervisionado***: o modelo é treinado em um conjunto de dados rotulados, ou seja, dados em que a resposta correta é conhecida (labels), relacionando os features com os labels para generalizar predições.
- ***Aprendizado Não Supervisionado***: os dados agora não tem labels (saídas desejadas), ou seja, não são rotulados, então, geralmente é utilizada para identificar padrões, como clustering ou redução de dimensionalidade.
- ***Aprendizado por Reforço***: esse tipo de aprendizado envolve um agente que interage com um ambiente e aprende a tomar ações que maximizam uma recompensa, geralmente utilizado em jogos.


7. ***O que é rede neural artificial e como ela funciona?***

Uma rede neural é um modelo computacional inspirado no cérebro humano, composto por camadas interconectadas de "neurônios" artificiais que processam informações para tarefas como reconhecimento de padrões, classificação e previsão, através do aprendizado de padrões em dados. Possui camadas de neurônios, que podem ser de entrada, saídas ou escondidas (Deep Learning) com parâmetros matemáticos como pesos e funções de ativação. Além disso, são construídas utilizando algoritmos de otimização, como o citado Backpropagation.

10. ***Como evitar vieses em conjuntos de dados e modelos de IA?***

O vies refere-se a uma tendência sistemática dos resultados de um modelo estarem inclinados em uma direção específica. Pode ocorrer quando por exemplo: coletamos muitos dados de uma parcela representativa da população, mas pouco de minorias; análise incorreta de dados, com a inserção de variáveis que têm pouca relação uma com a outra; não fazemos a limpeza dos dados, deixando colunas com valores inexistentes, isso afeta muito o modelo.

2. ***Um programa que faz cálculos matemáticos básicos (+, -, *, /) - utiliza as operações aritméticas do Python para realizar cálculos inseridos pelo usuário.***

````python
# Função Calculadora que pede aos usuários dois números e uma operação.
def calculator():
    print("Calculadora Simples")
    print("Operações disponíveis: +, -, *, /")
    print("_______________________________________")

    #Loop para calcular indefinidamente.
    while True:
        try:
            num1 = float(input("Digite o primeiro número: "))
            operator = input("Digite a operação (+, -, *, /): ")
            num2 = float(input("Digite o segundo número: "))

            # Cálculo conforme o operador inserido.
            if operator == '+':
                resultado = num1 + num2
            elif operator == '-':
                resultado = num1 - num2
            elif operator == '*':
                resultado = num1 * num2
            elif operator == '/':
                resultado = num1 / num2
            else:
                print("Operação inválida. Tente novamente.")
                continue

            print("Resultado:", resultado)

            # Se o usuário não deseja continuar, o loop se encerra.

            novamente = input("Deseja fazer outro cálculo? (s/n): ")
            if novamente.lower() != 's':
                print("Encerrando a calculadora.")
                break

        # Código preparado para erros de entrada inválida e divisões por  0.
        except ValueError:
            print("Entrada inválida. Certifique-se de inserir números válidos.")
        except ZeroDivisionError:
            print("Erro: Divisão por zero não é permitida.")

calculator()
````

3. ***Um programa que gera números aleatórios - importa e utiliza a biblioteca random para gerar números aleatórios.***

````python
import random

# A função abaixo gera n números entre li (limite inferior) e ls (limite superior)
def random_numbers(n, li, ls):
    random_num = []   # Lista vazia para armazenar os números aleatórios
    for _ in range(n):
        # Os n números aleatórios do intervalo [li,ls] são adicionados a random_num
        random_num.append(random.randint(li, ls))
    return random_num

# Solicitar os números, passar pela função e retornar printar a lista deles:
n = int(input("Digite a quantidade de números aleatórios desejados: "))
li = int(input("Limite inferior para os números aleatórios: "))
ls = int(input("Limite superior para os números aleatórios: "))

numeros_aleatorios = random_numbers(n, li, ls)

print("Os números são:", numeros_aleatorios)
````

4. ***Um programa que faz a média de uma lista de números - permite que o usuário insira vários números e utiliza loop e soma para calcular a média.***

````python
numeros = []

# Loop para solicitar os números ao usuário.
while True:
    entrada = input("Digite um número para ser adicionado à lista para o cálculo da média: ")
    try:
        numero = float(entrada)
        numeros.append(numero)
    except ValueError:
        print("Entrada inválida.")
        continue

    new = input("Deseja adicionar mais algum número? (s/n): ")
    if new.lower() != 's':
        print("Calculando a média.")
        break

# Media calculada com soma dividido pela quantidade de números (Usando numpy, a média é calculada diretamente com a função mean())
media = sum(numeros) / len(numeros)

print("Números inseridos:", numeros)
print("Média:", media)
````

5.  ***Um programa que conta a frequência de palavras em um texto - lê um texto, separa em palavras e utiliza um dicionário para contar frequência.***


````python
def count_frequency(texto):
    palavras = texto.split()    # Separa o texto com base nos espaços
    frequency = {}   # Dicionário para armazenar a palavra com "key" e frequência como "value"

    for i in palavras:
        i = i.lower()  # Transforma a palavra em minúsculas para evitar diferenciação
        if i in frequency:    # Se a palavra estiver em frequency, sua frequência aumenta  em 1
            frequency[i] += 1
        else:                 # Caso contrário, a frequência  é setada em 1, pois apareceu pela primeira vez
            frequency[i] = 1

    return frequency

texto = input("Digite o texto: ")
f = count_frequency(texto)    # f é um dict que armazena as palavra como "keys" e frequências como "values"

print("Frequência de palavras:")
for palavra, frequencia in f.items():   # Percorre os "keys" e "values" simultaneamente
    print(f"'{palavra}': {frequencia}")
````

6. ***Um programa que verifica se uma palavra é um palíndromo - checa se a string é igual de trás para frente.***

````python
# Função para checar palíndromo.
def palindromo(palavra):
    palavra = palavra.lower()  # Transforma a palavra em minúsculas para evitar diferenciação entre maiúsculas e minúsculas
    return palavra == palavra[::-1] # palavra[::-1] percorre a string toda de trás para frente, retornando a string ao contrário
    # Retorna "true" se a condição for verdadeira e "false" se não for

word = input("Digite uma palavra: ")

if palindromo(word):
    print("É um palíndromo.")
else:
    print("Não é um palíndromo.")
````

7. ***Um programa que adivinha um número aleatório escolhido pelo usuário - gera um número aleatório e permite que o usuário tente acertar.***

````python
import random   # Biblioteca para gerar números aleatórios

# A função guess implementará um jogo para adivinhar o número aleatótio entre um limite inferior (li) e um limite superior (ls)
def guess(li, ls):
    random_num = random.randint(li, ls)   # A função randint retorna uma número aleatório entre os limites do argumento
    tentativas = 0

    while True:
        palpite = int(input("Digite seu palpite: "))
        tentativas += 1   # Guarda o número de tentativas

        # Indução ao acerto:
        if palpite < random_num:
            print("Tente um número maior.")
        elif palpite > random_num:
            print("Tente um número menor.")
        else:
            print(f"CORRETO! Você adivinhou o número {random_num} em {tentativas} tentativas.")
            break

li = int(input("Digite o valor mínimo do número aleatório: "))
ls = int(input("Digite o valor máximo do número aleatório: "))

guess(li, ls)
````

8. ***Importar a base de dados Iris do scikit-learn e exibir em um dataframe usando Pandas - importa pandas e sklearn.datasets, carrega o Iris dataset e cria um dataframe a partir dos dados.***

````python
# Importar módulos utilizados
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris    # Importar o dataset

iris_data = load_iris()   # Salvar em "iris_data"

# O iris_data é um dict que contêm os dados no key 'data' e o nome das colunas em 'feature_names'
iris_df = pd.DataFrame(data = iris_data['data'], columns = iris_data['feature_names'])

# Está faltando só a atribuição da iris, prevista no key 'target'. Para isso, criemos uma nova coluna
iris_df['Iris_type'] = iris_data['target']

# Os nomes das iris estão com atribuições numéricas, vamos mudar isso
iris_df['Iris_name'] = iris_df['Iris_type'].apply(lambda x: 'sentosa' if x == 0 else ('versicolor' if x == 1 else 'virginica'))

print(iris_df['Iris_name'].value_counts())
iris_df.head()    # Mostra os primeiros 5 dados
````

10. ***Como normalizar colunas de dados (em um dataframe do pandas)? Ex.: fazer com que todos os valores estejam na faixa 0 a 1?***


Sendo $value_{max}$ o valor máximo e $value_{min}$ o valor mínimo de uma lista, podemos normalizar um valor ($value$) dessa lista da seguinte maneira:

$ normalized = \frac{value - value_{min}}{value_{max} - value_{min}} $

Ou, sendo $value_{mean}$ a média e $value_{std}$ desvio padrão:

$ normalized = \frac{value - value_{mean}}{value_{std}} $

Vamos normalizar os dados de _alcohol_ do _wine dataframe_:

````python
normalized_alcohol = wine_df['alcohol'].apply(lambda x: (x - wine_df['alcohol'].min())/(wine_df['alcohol'].max() - wine_df['alcohol'].min()))

df_normalized = pd.DataFrame({'Normalized Alcohol': normalized_alcohol, 'Alcohol':wine_df['alcohol']})
df_normalized
````

````python
# Teste se a normalização deu certo
maxValueIndex = df_normalized.idxmax()
minValueIndex = df_normalized.idxmin()

print("Index dos valores máximos e mínimos de cada coluna:")
print(maxValueIndex)
print(minValueIndex)
````

````python
print("Valores máximos: ")
print(f"Não normalizado: {df_normalized['Alcohol'][8]}, normalizado: {df_normalized['Normalized Alcohol'][8]}")
print("Valores mínimos: ")
print(f"Não normalizado: {df_normalized['Alcohol'][115]}, normalizado: {df_normalized['Normalized Alcohol'][115]}")
````
